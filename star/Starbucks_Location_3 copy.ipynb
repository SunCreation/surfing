# 3. 데이터 전처리

### 서울시 스타벅스 매장 목록, 인구, 사업체 데이터에 시군구명, 시군구코드 추가

```python
# 서울시 스타벅스 매장 목록이 담긴 엑셀 파일 불러오기
import pandas as pd
seoul_starbucks = pd.read_excel('seoul_starbucks_list.xlsx', header=0)
seoul_starbucks.head()
```

```python
# 스타벅스 주소 정보에서 시군구명 추출
sgg_names = []
for address in seoul_starbucks['주소']:
    sgg = address.split()[1]
    sgg_names.append(sgg)
seoul_starbucks['시군구명'] = sgg_names
seoul_starbucks.head()
```

```python
# 엑셀로 저장
seoul_starbucks.to_excel('seoul_starbucks_list.xlsx', index=False)
```
