## 2. 서울시 구별 인구 통계 가져오기

서울시 주민등록인구 통계
서울시 사업체 현황

두개 데이터 가져오기

```python
# 라이브러리
import pandas as pd
```

```python
# 서울시 구별 주민등록 인구 통계_영어로 바꿔주기
sgg_pop_df = pd.read_csv('report.txt', sep='\t', header=2)
columns = {
        '기간': 'GIGAN',
    '자치구': 'JACHIGU',
    '계': 'GYE_1',
    '계.1': 'GYE_2',
    '계.2': 'GYE_3',
    '남자': 'NAMJA_1',
    '남자.1': 'NAMJA_2',
    '남자.2': 'NAMJA_3',
    '여자': 'YEOJA_1',
    '여자.1': 'YEOJA_2',
    '여자.2': 'YEOJA_3',
    '세대': 'SEDAE',
    '세대당인구': 'SEDAEDANGINGU',
    '65세이상고령자': 'N_65SEISANGGORYEONGJA'
}
sgg_pop_df.rename(columns= columns, inplace = True)
```

```python
# head()
sgg_pop_df.head()
```

```python
# info()
sgg_pop_df.info()
```

```python
# tail()
sgg_pop_df.tail()
```

```python
# 필요없는 데이터 제거 '합계'
condition = sgg_pop_df['JACHIGU'] != '합계'
sgg_pop_df_selected = sgg_pop_df[condition]
sgg_pop_df_selected.head()
```

```python
# info()
sgg_pop_df_selected.info()
```

```python
# 분석에 필요한 칼럼 선택 'JACHIGU', 'GYE_1'
columns = ['JACHIGU', 'GYE_1']
sgg_pop_df_final = sgg_pop_df_selected[columns]
```

```python
# head()
sgg_pop_df_final.head()
```

```python
# 데이터프레임의 칼럼명 변경 '시군구명','주민등록인구'
sgg_pop_df_final.columns = ['시군구명','주민등록인구']
sgg_pop_df_final.info()
sgg_pop_df_final.head()
```

```python
# 엑셀로 저장하기
sgg_pop_df_final.to_excel('sgg_pop.xlsx', index=False)
```

```python
# 서울시 동별 사업체 현황 통계 데이터 가져오기
sgg_biz_df = pd.read_csv('report.txt', sep='\t', header=2)
sgg_biz_df.head()
```

```python
# 컬럼명 바꾸기

columns = {
    '기간': 'GIGAN',
    '자치구': 'JACHIGU',
    '동': 'DONG',
    '사업체수': 'SAEOPCHESU_1',
    '여성대표자': 'YEOSEONGDAEPYOJA',
    '계': 'GYE',
    '남': 'NAM',
    '여': 'YEO'
}
sgg_biz_df.rename(columns= columns, inplace = True)
sgg_biz_df.head(10)
```

```python
# info()
sgg_biz_df.info()
```

```python
# 시군구별 사업체 현황 데이터 추출 'DONG' == '소계'
condition = sgg_biz_df['DONG'] == '소계'
sgg_biz_df_selected = sgg_biz_df[condition]
sgg_biz_df_selected.head(10)
```

```python
# 필요없는 칼럼 제거
columns = ['JACHIGU', 'GYE', 'SAEOPCHESU_1']
sgg_biz_df_final = sgg_biz_df_selected [columns]
sgg_biz_df_final.columns = ['시군구명', '종사자수','사업체수']
sgg_biz_df_final.head()
```

```python
# 데이터프레임의 인덱스 초기화
sgg_biz_df_final[['시군구명','사업체수']].to_excel('sgg_biz.xlsx',index=False)
```
